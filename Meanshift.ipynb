```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv("Mall_Customers.csv")
```

```python
X=Dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import MeanShift
y_Mshift = MeanShift(bandwidth=25)
labels = y_Mshift.fit_predict(X)
```

```python
Mshift=y_Mshift.fit_predict(X)
```

```python
Mshift
```

```python
supervised=pd.DataFrame(Dataset)
supervised['cluster_group3']=Mshift
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
centroids=y_Mshift.cluster_centers_
```
